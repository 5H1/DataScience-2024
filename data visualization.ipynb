```python
#step 2 a. performing Box plot on Volume column and observing outliers and inter quartile ranges
ds.boxplot(column='Volume',vert=False)
```

```python
#step 2 a. performing Box plot on Volume column and observing outliers and inter quartile ranges
ds.boxplot(column='Volume',vert=False)
import numpy as np
q1 = np.percentile(ds["Volume"],25)
print("25th percentile",q1)
q3 = np.percentile(ds["Volume"],75)
print("75th percentile",q3)
iqr = q3-q1
print("Inter quartile range:",iqr)
UW = q3 + (1.5*iqr)
print("Upper whisker length:",UW)
print("number of outliers greater than upper whisker:",len(ds["Volume"][ds["Volume"] > UW]))

#Observations:

#1.Most of the data points lies on the left side.

#2.The box plot is present over left side therefore the assumption is most of the data points lies on left side and thus it is positively skewed.

#3.25th percentile is 3.0

#4.75th percentile that is Q3=6.0

#5.Inter quartile range is 3.0.

#6.Outliers are lying above upper whisker length.
```
